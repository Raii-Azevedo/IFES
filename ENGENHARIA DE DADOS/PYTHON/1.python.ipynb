O que é Python e para que serve?

Olá, Guerreiro Jedi! Nesta unidade trataremos a respeito da linguagem de programação Python, compreendendo as suas principais características, bem como, a sua aplicação prática no mercado dentro de seus mais variados segmentos. Aperte os cintos pois a jornada será empolgante!

O Python surgiu na década de 1990 e, embora já seja uma linguagem com bastante maturidade, ela só começou a ganhar notoriedade da década de 2010 em diante - momento em que a Inteligência Artificial (I.A.) começou a se tornar cada vez mais presente nas mais diversas aplicações, sejam elas web, mobile ou desktop.

Como o Python, desde sua concepção já trazia uma estrutura robusta voltada à I.A. e aprendizado de máquina por meio de bibliotecas específicas, a linguagem acabou por atender perfeitamente às necessidades de programadores que precisavam desenvolver aplicações inteligentes. A estrutura do Python também é bastante expansível, ou seja, a linguagem permite certa modularidade nas aplicações.

Obviamente que este não é o único motivo pelo qual o Python ganha destaque no cenário da programação mundial. A linguagem é muito conhecida também por sua velocidade de execução - principalmente quando processada em ambiente local.

De acordo com a documentação oficial da linguagem, o “Python é uma linguagem de programação que permite trabalhar rapidamente e integrar os sistemas de forma mais eficaz” (Python.org, 2022).

Mas nem tudo são flores, pequeno gafanhoto. O Python também, assim como qualquer outra linguagem, tem seus pontos negativos conforme veremos ao longo do material e das aulas em vídeo. Isso não tira seu mérito da linguagem tampouco dos programadores que a adotam como padrão de desenvolvimento, afinal, as vantagens da linguagem sobressaem seus déficits.

Em suma, mas não exclusivamente, Python é amplamente utilizada para:

Criação de aplicações baseadas em Inteligência Artificial;
Criação de aplicações que requeiram aprendizado de máquina, ou seja, softwares que aprendam padrões com o passar do tempo, de forma a tomar decisões guiadas por decisões assertivas do passado;
Ciência e Análise de dados, provendo cenários preditivos para tomada de decisão a nível corporativo ou governamental;
Sites, lojas virtuais, sistemas que funcionem no padrão SAAS (Software As A Service) e qualquer outro tipo de aplicação que seja executada em browser (navegador);
Sistemas de automação residencial e/ou industrial;
Softwares embarcados, ou seja, aplicações que interajam diretamente com o hardware - os sistemas de automação são um bom exemplo disso;
Aplicativos Mobile (embora não seja a linguagem mais adequada para isso e, portanto, requeira alguns frameworks e camadas adicionais para funcionar da maneira esperada).
O Python consegue estar presente em basicamente qualquer área da tecnologia que envolva software ou software+hardware, cabendo única e exclusivamente ao programador optar por seu uso dentro do projeto que está sendo desenvolvido.

Fazendo o download do Python

Você deverá acessar o site oficial, https://www.python.org, e acessar a guia de downloads. Recomendo que você verifique a versão atual e procure sempre instalar a versão “stable” (estável). Como a linguagem está em constante atualização, novas funções, recursos, bibliotecas, patches de segurança e outras manutenções, são adicionadas periodicamente.

A cada pacote de atualização adicionado, uma nova sequência é adicionada ao pacote padrão do Python. Quando o pacote de atualizações é extenso e modifica de maneira considerável a estrutura da linguagem, é lançada uma nova versão.

Novas versões podem ser instáveis e conter pequenos pontos de ajustes que passaram despercebidos pela comunidade de desenvolvedores, ou seja, falhas e erros podem ocorrer sem aviso prévio e com um agravante: sem uma solução pré-definida.

Desta forma, recomendo sempre que você utilize uma versão estável, ou seja, que já tenha sido validada e passado no “crivo” dos desenvolvedores, resolvendo todos os problemas (ou a maioria, ao menos) que aquela versão tenha.

Escolher uma versão em desenvolvimento, ainda que ela ofereça algumas vantagens como novos recursos e até mesmo a resolução de erros de versões anteriores, é um passo arriscado, uma vez que a versão em desenvolvimento pode gerar uma série de incompatibilidades que acabarão sendo mais danosas que o uso de uma versão com menos recursos. Em suma: use sempre a versão estável se quiser diminuir a dor de cabeça!

Entendendo a arquitetura do Python

O Python é uma linguagem fracamente tipada (tipagem dinâmica) e interpretada. Esses dois termos se referem à forma como a linguagem é lida pelo processador (ou interpretador) de uma máquina. A “tipagem” se refere aos tipos de dados que uma variável pode assumir; já o termo “interpretada”, significa que a linguagem, via de regra, será executada por uma aplicação que interprete o código - motivo pelo qual ela tende a ser mais rápida que outras linguagens, ao menos com processos triviais que não requeiram loops altos (veremos o que são loops / laços de repetição um pouco mais adiante, combinado?).

 

Explorando as variáveis

Variáveis, em programação, nada mais são que espaços na memória que podem ou não alocar determinados valores. Este “ou não”, refere-se a linguagens de tipagem forte, como o C, por exemplo. Ao se trabalhar com uma linguagem fortemente tipada, é necessário informar à memória que uma determinada variável terá um tipo específico (texto, número inteiro, número decimal, etc). A esse processo dá-se o nome de “declaração de variável”.

Embora não seja este o foco desta disciplina, é importante que você compreenda o comportamento de outras linguagens, de forma a entender uma das várias vantagens do Python. Variáveis em C teriam uma estrutura similar a essa demonstrada a seguir:

char nome;

int idade;

Essa declaração de variáveis, faz com que um espaço seja reservado na memória, possibilitando, em momento propício, alocar um dado no espaço reservado. Um detalhe importante sobre linguagens fortemente tipadas como o C, é a restrição de modificação de tipos. Se uma variável foi definida como sendo do tipo “inteiro”, ela será deste tipo até a finalização do programa.

Já com linguagens como o Python, fracamente tipadas, não há necessidade de se reservar um espaço na memória, ou seja, a declaração de variáveis é desnecessária. Por possuir uma tipagem dinâmica, uma mesma variável pode armazenar dados diversos, por exemplo, textos em um momento e números em outro momento, conforme demonstrado a seguir:

valor = “Python”

valor = 10

Em um primeiro momento, a variável “valor” armazenava o texto “Python” e já na linha seguinte, ela passou a armazenar o número inteiro 10. Essa tipagem dinâmica facilita bastante o processo de desenvolvimento e, embora isso possa abrir brechas para erros ou falhas, o programador acaba tendo mais ferramentas de trabalho durante o processo de desenvolvimento.

As variáveis podem armazenar valores diversos: textos, caracteres isolados, números inteiros, números com casas decimais, etc. Entenda que uma variável é uma “caixa vazia e pequena” na qual pode-se armazenar o que quiser, porém, devido ao seu tamanho, só é possível armazenar uma informação de cada vez, logo, se um valor diferente for armazenado em uma variável, o valor anterior será perdido conforme você pode verificar a seguir:

idade = 20

idade = 30

O resultado de execução deste código será o seguinte:

a. A variável idade assume o valor 20

b. A variável idade assume o valor 30, removendo o valor 20

c. O valor 20 perde-se definitivamente

Explorando as constantes

Diferente de outras linguagens, o Python, por ter uma tipagem dinâmica, não utiliza o conceito de constantes, ou seja, ao se determinar uma variável é possível modificar seu valor a qualquer momento. É importante salientar que isso é válido até a versão atual da linguagem e que a qualquer momento essa informação pode ser modificada, caso a comunidade de desenvolvedores lance uma versão que contraponha essa regra.

Caso você opte por determinar uma variável cujo valor NUNCA deva ser alterado, você pode escrevê-la em maiúsculo. Isso não impedirá sua modificação, mas dará um alerta visual a quem estiver manipulando o código, sinalizando que aquela variável é diferente das demais.

 

Regras básicas de variáveis

Basicamente, as variáveis são elementos capazes de armazenar valores e podem ter o nome que você desejar. Algumas regras, entretanto, devem ser respeitas, como:

Variáveis não podem iniciar com números
Variáveis não podem ter espaço
Variáveis não podem ter caracteres especiais (!, @, #, $, etc)
Para nomes de variáveis que contenham, conceitualmente, duas ou mais palavras, utilizamos o conceito de Camel Case, escrevendo as iniciais, a partir da segunda palavra, em maiúsculo, por exemplo, uma variável para armazenar um nome do usuário, poderia ser descrita como: nomeDoUsuario. Perceba que a primeira palavra está em minúsculo e as iniciais das palavras seguintes, em maiúsculo. Adicionalmente, a palavra “usuário” não tem acentuação, pois isso iria ferir uma das regras de variáveis.
